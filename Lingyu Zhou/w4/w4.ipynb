```python
import pandas as pd
from tqdm import tqdm
```

```python
df=pd.read_csv("../w3/buyers/buyers_cleaned.csv")
df=df.fillna("")
```

```python
bnl=["buyer1name",	"buyer2name",	"buyer3name"]
bil=["buyer1FTindex",	"buyer2FTindex",	"buyer3FTindex"]
```

```python
# func. def.

## discarded cuz need index:
# def isin(strg, lst):
#     '''strg: target string\n\nlst: iterable'''
#     return any(set([strg in x for x in lst]))

def nestedIn(strg,lst):
    '''strg: target string\n\nlst: iterable'''
    idx=-1
    if strg.strip()=="":
        return -2
    for i in range(len(lst)):
        if strg in lst[i]:
            idx=i
    return idx

def getAllFact(lim=["buyer1name",	"buyer2name",	"buyer3name"]):
    l=[]
    for i in lim:
        l.extend(list(df[i]))
    return set([x for x in l if x!=""])

def assign(buyer,result):
    global df
    for idx, row in df.iterrows():
        for buyer_col, bi_col in zip(bnl, bil):
            if row[buyer_col] == buyer:
                df.loc[idx, bi_col] = result

def kwQuery(kw, country:list=None,bw=False):
    global df
    if country!=None:
        tdf=df[df.Country.apply(lambda x:x in country)]
    else:
        tdf=df
    if bw==False:
        return tdf[tdf[bnl].apply(lambda r: r.str.contains(kw)).any(axis=1)]
    else:
        return tdf[(tdf[bnl].apply(lambda r: r.str.contains(int(kw))).any(axis=1))
                  |
                  (tdf[bnl].apply(lambda r: r.str.contains(kw)).any(axis=1))]
```

```python
# some forgotten preprocessings
for i in bnl:
    df[i]=df[i].str.replace("  "," ").str.strip()

# creating 3 new cols:
for i in range(3):
    colname="buyer"+str(i+1)+"FTindex"
    df[colname]=""
```

```python
# get score dict (len=150):
with open("scores.txt") as sc:
    sc=sc.read().split("\n")

sc=[x.strip() for x in sc]
sc={x.rsplit(" ",maxsplit=1)[0].strip().lower(): int(x.rsplit(" ",maxsplit=1)[1]) for x in sc}
sck=list(sc.keys())
sc
```

```python
# assign scores:
failed=[]
for i,r in df.iterrows():
    for j in zip(bnl,bil):
        if (indx:=nestedIn(r[j[0]] ,sck))>-1:
            # print(indx, sck[indx])
            df.loc[i,j[1]]=sc[sck[indx]]
        elif indx==-2:
            df.loc[i,j[1]]=-2
        else:
            df.loc[i,j[1]]=-1
            failed.append(r[j[0]])

len(set(failed))
```

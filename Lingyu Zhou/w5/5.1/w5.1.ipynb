```python
import pandas as pd
import re
```

```python
def padList(l,leng,padding=""):
    assert len(l)<=leng
    return l+([padding]*(leng-len(l)))

def parseFloatStr(s):
    if s!='nan':
        return str(round(float(s)*100,4))+"%"
    else:
        return ''
```

```python
with open("otr.txt","r") as otr:
    otr=otr.read().split("#")
otr=[x.replace("\t"," ").replace("\n"," ").replace(",","").strip() for x in otr if x!=""]
# otr
rates={}
for i in range(len(otr)):
    item=otr[i]
    if item[0].isalpha():
        rates[item]=otr[i+1].split("@")
    else:
        continue
# rates
```

```python
typeList=["Volume_pairs","Customer_Target",	"NB_Requested",	"Orig_Confirmed"]
df=dict(zip(typeList,[[],[],[],[]]))
for i in rates:
    for j in range(len(rates[i])):
        op=re.sub(r'\s+', ' ', rates[i][j]).split(" ")
        op=[x.replace("-1","") for x in op if x.strip()!=""]
        df[typeList[j]].extend(padList(op,48))

df
```

```python
df=pd.DataFrame(df)
df["factorycode"]=[x for x in list(rates.keys()) for _ in range(48)]
df["year"]=[x for x in range(2021,2025) for _ in range(12)]*6
df["month"]=[i for i in range(1,13)]*24
df=df[list(df.columns[-3:])+list(df.columns[:-3])]
df
```

```python
df.to_csv("factory_on_time.csv",index=False)
```

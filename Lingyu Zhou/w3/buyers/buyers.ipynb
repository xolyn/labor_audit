````python
import pandas as pd
````

````python
df=pd.read_csv("buyers.csv")
df=df.drop(["Unnamed: 7"],axis=1).fillna("")
df[["buyer1name","buyer2name","buyer3name"]]=df[["buyer1name","buyer2name","buyer3name"]].apply(lambda x: x.str.lower())
bnl=["buyer1name","buyer2name","buyer3name"]
bwl=['buyer1BW', 'buyer2BW', 'buyer3BW']
allCountries=list(df.Country.unique())
````

````python
def cleanStrList(los):
    los=sorted([x.replace("  "," ").strip() for x in los])
    return [x.lower() for x in los if x!=""]

def getAllBuyer(country):
    l=[list(df[df.Country==country][x].unique()) for x in bnl]
    fl=list(set([item for sublist in l for item in sublist]))
    return [x for x in fl if x!=""]

def getUnmatched(country):
    global df
    buyer_names = df.loc[df.Country==country,bnl]
    buyer_bw = df.loc[df.Country==country,bwl]
    return sorted(list(set([name
            for names_row, bw_row in zip(buyer_names.values, buyer_bw.values)
            for name, bw in zip(names_row, bw_row)
            if bw == 0])))

def kwQuery(kw, country:list=None,bw=False):
    global df
    if country!=None:
        tdf=df[df.Country.apply(lambda x:x in country)]
    else:
        tdf=df
    if bw==False:
        return tdf[tdf[bnl].apply(lambda r: r.str.contains(kw)).any(axis=1)]
    else:
        return tdf[(tdf[bnl].apply(lambda r: r.str.contains(int(kw))).any(axis=1))
                  |
                  (tdf[bnl].apply(lambda r: r.str.contains(kw)).any(axis=1))]
    
def assign(buyer,result,country:list):
    global df
    for index, row in df[df.Country.apply(lambda x:x in country)].iterrows():
        for buyer_col, bw_col in zip(bnl, bwl):
            if row[buyer_col] == buyer:
                df.loc[index, bw_col] = result

def checkBuyer(country:str, factlist:list):
    global df
    for i,j in df[df.Country==country].iterrows():
        for k in bnl:
            corrfield=k.replace("name","BW")
            if j[k]!="":
                result=1 if True in set([j[k] in x for x in factlist]) else 0
                df.loc[i,corrfield]=result
````

````python
# factory list define:
with open("./buyers.txt","r") as buyers:
    buyers=buyers.read().split("@")[1:]

# buyers
indo=cleanStrList(buyers[0].split("\n")[1:])

camb=cleanStrList(buyers[1].split("\n")[1:])

jord=cleanStrList(buyers[2].split("\n")[1:])

nica=cleanStrList(buyers[3].split("\n")[1:])

bang=cleanStrList(buyers[4].split("\n")[1:])
````

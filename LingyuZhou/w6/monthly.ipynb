| var | desc | req. |
| ----------- | -------------- |------------------ |
| 1 style_mix     | count the number of unique styles in each year-month for each factory.                                                                                                                           | This is to be calculated from the variable "Style" for each factory in data2 dataset.        |
| 2 order_batches | count the number of order batches in each year-month for each factory                                                                                                                            | This is to be calculated from the variable "Order#" for each factory in data2 dataset.       |
| 3 order_quant   | the total order quantity (in paires or pieces) in each year-month for each factory                                                                                                               | This is to be calculated from the variable "Quantity" for each factory in data2 dataset.     |
| 7 lead_timeim   | the weighted ideal lead time in each month for each factory; the weight for each lead time is the percentage of "quantity" for the particular order among the total order quantity in the month. | calculation: (lead_timei(1) x weight1) + (lead_timei(2) x weight2)…..                        |
| 8 lead_timenm   | the weighted buyer lead time in each month for each factory.                                                                                                                                     | weight and calculation are the same as above                                                 |
| 9 lead_timecm   | the weighted factory confirmed lead time in each month for each factory                                                                                                                          | weight and calculation are the same as above                                                 |
| ------              | **The following three variables are to be calculated for each order in data2 (not for each month)**                                                                                      |-----
| 4 lead_timei    | the number of days from release of order to ideal customer target ontime delivery for each order.                                                                                                | Calculation with variables in data2: "Customer Target XFD (CTXFD)"-"Release Date"            |
| 5 lead_timen    | the number of days from release of order to buyer requested delivery date for each order.                                                                                                        | Calculation with variables in data2: "NB Requested XFD (NBRXFD)"-"Release Date"              |
| 6 lead_timec    | the number of days from release of order to factory confirmed delivery date for each order.                                                                                                      | Calculation with variables in data2: "Orig Fact Cfm XFD (OCXFD)"-"Release Date"              |

# Round1:
`"style_mix", "order_batches", "order_quant"`:

````python
for i in fact:
    print("processing",i)
    dft=pd.read_csv(i,thousands=",")
    dft.loc[dft[dft["SI Month"]=="2022/00"].index,"SI Month"]="2021/12"

    # style_mix:
    assign(i.split(".")[0],"style_mix",dft.groupby("SI Month")["Style"].nunique())
    # order_batches:
    assign(i.split(".")[0],"order_batches",dft.groupby("SI Month")["Order#"].nunique())
    # order_quant:
    assign(i.split(".")[0],"order_quant",dft.groupby("SI Month")["Quantity"].sum())

    print(i,"done processing")
    del dft
````

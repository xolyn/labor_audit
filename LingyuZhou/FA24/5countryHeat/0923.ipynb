```python
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
plt.rcParams["font.family"]="Arial"
plt.rcParams.update({'font.size': 25})
```

```python
k=pd.read_excel('D:/OneDrive - Cornell University/JselZJOB/24SU/w14/Karachi.xlsx',sheet_name='Weather Data')
p=pd.read_excel('D:/OneDrive - Cornell University/JselZJOB/24SU/w14/Phnom Penh.xlsx',sheet_name='Weather Data')
d=pd.read_excel('D:/OneDrive - Cornell University/JselZJOB/24SU/w14/Dhaka.xlsx',sheet_name='Weather Data')
# c=pd.read_excel('D:/OneDrive - Cornell University/JselZJOB/24SU/w14/Chattogram.xlsx',sheet_name='Weather Data')
h=pd.read_excel('D:/OneDrive - Cornell University/JselZJOB/24SU/w14/Ho Chi Minh.xlsx',sheet_name='Weather Data')

k=k.loc[k.datetime.str[:4].astype(int)>=2010, ['datetime','tempmax']]
p=p.loc[p.datetime.str[:4].astype(int)>=2010, ['datetime','tempmax']]
d=d.loc[d.datetime.str[:4].astype(int)>=2010, ['datetime','tempmax']]
h=h.loc[h.datetime.str[:4].astype(int)>=2010, ['datetime','tempmax']]
```

```python
plt.close("all")
```

```python
names={'k':'Karachi','p':'Phnom Penh','d':'Dhaka','h':'Ho Chi Minh'}
dataframes = [k, p, d, h]
titles = list(names.values())

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(32,18))
# fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(35,20))
ax = axes.ravel()

for i, df in enumerate(dataframes[2:]):
    ax[i].plot(df.index, df['tempmax'], color='orange',linewidth=1.5)
    
    condition = df['tempmax'] > 35
    ax[i].scatter(df.index[condition], df['tempmax'][condition], color='red',linewidth=5)
    
    ax[i].axhline(y=35, color='red', linestyle='-', linewidth=2.5, label='baseline (35C)')
    ax[i].set_title(titles[i+2],size=30)
    ax[i].set_xlabel('Date')
    ax[i].set_ylabel('Maximum Temperature')
    ax[i].legend()
    
plt.tight_layout()

# Show the plot
plt.show()
```

```python
names = {'k': 'Karachi', 'p': 'Phnom Penh', 'd': 'Dhaka', 'h': 'Ho Chi Minh'}
dataframes = [k, p, d, h]  # Assume 'k', 'p', 'd', 'h' are previously defined dataframes
titles = list(names.values())

# Creating a starting date
start_date = pd.to_datetime('2010-01-01')

# Creating date ranges for each dataframe and setting them as new index
for df in dataframes:
    date_range = pd.date_range(start=start_date, periods=len(df), freq='D')
    df.set_index(date_range, inplace=True)

# Plotting
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(32, 18))
ax = axes.ravel()

for i, df in enumerate(dataframes[:2]):
    ax[i].plot(df.index, df['tempmax'], color='orange', linewidth=1.5)
    
    condition = df['tempmax'] > 35
    ax[i].scatter(df.index[condition], df['tempmax'][condition], color='red', linewidth=5)
    
    ax[i].axhline(y=35, color='red', linestyle='-', linewidth=2.5, label='baseline (35C)')
    ax[i].set_title(titles[i], size=30)
    ax[i].set_xlabel('Date')
    ax[i].set_ylabel('Maximum Temperature')
    ax[i].legend()
    ax[i].grid(True)  # Optional: adds grid for better visibility

plt.tight_layout()

# Show the plot
plt.show()
```
